**Performing a boxplot of the total bill by day.**





```python
import pandas as pd
import matplotlib.pyplot as plt
Data =pd.read_csv('/content/tips.csv')
Data.boxplot(by ='day', column =['total_bill'], grid=False)
plt.ylabel('Total Bill')
plt.xlabel('Days')
plt.show()
```

**Creating a boxplot using Seaborn.**

```python
import seaborn
seaborn.set(style='whitegrid')
tip = pandas.read_csv('/content/tips.csv')
seaborn.boxplot(x='day', y='tip', data=tip)
```
