**For Linear Regression Question**

Importing library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Adding Data  value of x and y values from the question

```python
x= [10,20,30,40,50,60,70,80,90]
y= [420,365,285,220,176,117,69,34,5]
```

  Performing the scatter plot

```python
plt.xlabel('Temperature in Celcius (x)')
plt.ylabel('Life time in hours (Y)')
plt.scatter(x, y)
plt.show()
```

**Performing linear regression on the data in the lists x and y.**

```python
from scipy import stats
B1, B0, r, p, std_err = stats.linregress(x, y)
print(B0,B1,r,std_err)
```

**Returns the value of the linear equation**

```python
def myfunc(x):
  return B0 + B1 * x
```

**Performing statsmodels module to fit a linear regression model.**

```python
import statsmodels.api as sm
mymodel = list(map(myfunc, x))
speed = myfunc(20)
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.xlabel('Temperature in Celcius (x)')
plt.ylabel('Life time in hours (Y)')
plt.show()
print(speed)
x = sm.add_constant(x)
x = np.array(x)
model = sm.OLS(y, x).fit()
print(model.summary())
```
